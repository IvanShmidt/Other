```python
# ENCRIPTION

ENCODING = 'utf-8'
st = "Привет. Я пошел!!!"

enc_str = bytearray(st, encoding=ENCODING)
rest_spaces = (width*height - len(BitArray(bytes=enc_str).bin)) / 8
st += int(rest_spaces) * " "
b_arr = bytearray(st, encoding=ENCODING)
bits_str = BitArray(bytes=b_arr).bin
bits_arr = [int(i) for i in bits_str]

original = Image.open('container.png')
width, height = original.size

if width*height == len(bits_arr):
    print('size matched.')

steg = Image.new('RGB', (width, height))

idx = 0
for i in range(width):
    for j in range(height):
        r, g, b = original.getpixel((i, j))
        if bits_arr[idx] == 0:
            r &= 254
        else:
            r |= 1
        steg.putpixel((i, j), (r, g, b))
        idx += 1
        
steg.save('encripted.png')
```

```python
def random_ints_with_sum_and_length(n, _len):
    # n -верхняя граница, можем брать только числа, которые в сумме будут меньше n
    # _len - минимальная длина массива на выходе, т.е. количество наших бит, которые нежно закодировать
    
    def random_ints_with_sum(n, seed):
        """
        Generate non-negative random integers summing to n.
        """
        random_gen = np.random.RandomState(seed)
        while n > 1:
            r = random_gen.randint(1, n)
            yield r
            n -= r
            
    seed = 1
    ints = list(random_ints_with_sum(n, seed))
    
    while len(ints) < _len:
        seed += np.random.randint(1, 10)
        ints = list(random_ints_with_sum(n, seed))
        
    return ints, seed
```
